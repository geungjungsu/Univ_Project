## 피처요약표 함수

```python
def feature_summary(df) :
  df_feature = pd.DataFrame()
  df_feature['타입'] = df.dtypes
  df_feature['결측치'] = df.isna().sum()
  df_feature['고유값'] = df.nunique()
  df_feature['중복치'] = df.duplicated().sum()
  df_feature['최소값'] = df.min()
  df_feature['최대값'] = df.max()
  df_feature['값1'] = df.iloc[0]
  df_feature['값2'] = df.iloc[1]
  df_feature['값3'] = df.iloc[2]
  return df_feature
```

### GRDP : GRDP_city_real

- 시도별
    - 서울특별시, 경기도, 경상남도, 제주특별자치도
- .1
    - 실질
        - 명목 : 물가 상승률을 고려하지 않은 경제적 가치
        - 실질 : 물가 변동을 반영하여 조정된 경제적 가치
        - 실질기여도 : 특정 경제 활동이 전체 GOP에 얼마나 기여했는지를 나타내는 지표, 실질적으로 계산

```python
# 데이터 불러오기
GRDP = pd.read_csv('/content/drive/MyDrive/미드 프로젝트_이정수/GRDP.csv',encoding='cp949')

# 데이터 확인
GRDP.head()
GRDP.shape
feature_summary(GRDP)
```

#### 전처리_GRDP

##### 시도별

```python
# 시도별 : 서울특별시, 경기도, 경상남도, 제주특별자치도

# 시도명 확인
GRDP['시도별'].unique()

# 특정 시도 필터
GRDP_city = GRDP[(GRDP['시도별'] == '서울특별시') | (GRDP['시도별'] == '경기도') | (GRDP['시도별'] == '경상남도') | (GRDP['시도별'] == '제주특별자치도') ]

# 다른 방법
# GRDP[GRDP['시도별'].isin(['서울특별시','경기도','경상남도','제주특별자치도'])]
```

##### 실질(.1)

```python
# 시도별, 경제활동별, .1 추출
GRDP_city_real = GRDP_city[[col for col in GRDP_city.columns if ('.1' in col) or (col in ['시도별','경제활동별'])]]
GRDP_city_real

# 변수명 변경 : .1 -> 실질로 변경
GRDP_city_real.columns = GRDP_city_real.columns.str.replace('.1','실질').str.replace(' p)','')

# 파일 생성
GRDP_city_real.to_csv('지역별_GRDP_TableauFile.csv',encoding='utf-8-sig',index=False)
```

### 실업률, 고용률

#### 전처리_실업률_고용률

```python
# 연도(기간) 확인
year_youth_employment.index # 2011-2023
unemployment_rate.index # 2013-2023


# 기간 통일 -> 2013년부터
new_year_youth_employment = year_youth_employment[year_youth_employment.index >= 2013]

# type 통일 -> index type : int64로 변경
unemployment_rate.index = unemployment_rate.index.astype('int64')

# 데이터 합치기
unemployment_employment = pd.merge(unemployment_rate, new_year_youth_employment, left_index=True, right_index=True, how='left')

# 파일 생성
unemployment_employment.to_csv('실업률_고용률_TableauFile.csv',encoding='utf-8-sig')
```

### 최저임금, 연도별 청년고용률

```python
# 연도(기간 확인)
minimum_wage.index # 2013-2023
new_year_youth_employment.index # 2013-2023

# type 통일
minimum_wage.index = minimum_wage.index.astype('int64')

# 데이터 합치기
wage_employment = pd.merge(minimum_wage, new_year_youth_employment, left_index=True, right_index=True, how='left')

# 파일 생성
wage_employment.to_csv('최저임금_고용률_TableauFile.csv',encoding='utf-8-sig')
```

### 시도연령별실업률, 일자리지원율

```python
# 연도(기간 확인)
unemployment_rate.index # 2013-2023
employment_support.index # 2015-2022

# 기간 통일 -> 2015년부터 2022년까지
unemployment_rate = unemployment_rate[(unemployment_rate.index >= 2015) & (unemployment_rate.index <= 2022)]

# 데이터 합치기
unemployment_employment_support = pd.merge(unemployment_rate, employment_support, left_index=True, right_index=True, how='left')

# 파일 생성
unemployment_employment_support.to_csv('실업률_일자리지원율_TableauFile.csv',encoding='utf-8-sig')
```
